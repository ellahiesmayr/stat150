# Bayes' rule

In this notebook we will explore the notion of conditional probability and Bayes' rule. To do this we will revisit the traditional example of medical tests. The question we are trying to answer is, what is the probability that the person is sick, given that the test was positive, when we know the proportion of people that are sick in the population, the probability that the test is positive, when the person is sick, and the probability that the test is positive, when the person is healthy. We know from Bayes' rule and the law of total probability that
$$\mathbb{P}(\text{person is sick }| \text{ test is positive }) = \frac{\mathbb{P}(\text{person is sick})\mathbb{P}(\text{test is positive} | \text{person is sick})}{\mathbb{P}(\text{test is positive})} = \frac{\mathbb{P}(\text{person is sick})\mathbb{P}(\text{test is positive} | \text{person is sick})}{\mathbb{P}(\text{person is sick})\mathbb{P}(\text{test is positive} | \text{person is sick}) + \mathbb{P}(\text{person is healthy})\mathbb{P}(\text{test is positive} | \text{person is healthy})}.$$

The following function does exactly the above calculation, and takes as input the three mentioned parameters.

```python
# this function calculates the probability that someone is sick 
# given that their test is positive, when
# the probability that the test is positive,
# when the person is sick, is p_pos_when_sick, and
# the probability that the test is positive,
# when the person is healthy, is p_pos_when_healthy,
# and the proportion of the population that is sick is p_sick.

def p_sick_when_pos(p_sick, p_pos_when_sick, p_pos_when_healthy):
    return ( p_pos_when_sick * p_sick )/(p_pos_when_sick * p_sick
                                                + p_pos_when_healthy * (1-p_sick))
```

We will now graph the probability that a person is sick, given that she tested positive, wich respect to the proportion of the population that is sick. In this first graph we keep the probability that the test is positive when the person is sick constant and loop through a few different values for the probability that a healthy person tests positive. In the second part we do exactly the opposite.

```python
# probability that the test is positive when the person is sick
p_pws = 0.99

plt.figure(figsize=(20, 8))
# Remove the plot frame lines.
ax = plt.subplot(111)    
ax.spines["top"].set_visible(False)      
ax.spines["right"].set_visible(False)    

# Ensure that the axis ticks only show up on the bottom and left of the plot.    
ax.get_xaxis().tick_bottom()    
ax.get_yaxis().tick_left()

plt.xlabel("Proportion of the population that is sick.")
plt.ylabel("Probability that a person is sick given that they test positive.")

for p_pwh in np.arange(0.001, 0.01, 0.001):
    x = np.arange(0.00001, 0.10, 0.00001)
    y = p_sick_when_pos(x, p_pws, p_pwh) 
    plt.plot(x,y, label = str(round(p_pwh,3)));
ax.legend(title = "Probability that a healthy persons test positive")
plt.title("Effectiveness of a test when the probability" + 
          " that a sick person tests positive is " + str(p_pws) +
         "for different values of the probability that a healthy person tests positive.");
```

```python
# probability the test is positive when the person is healthy
p_pwh = 0.01

plt.figure(figsize=(20, 8))
# Remove the plot frame lines.
ax = plt.subplot(111)    
ax.spines["top"].set_visible(False)      
ax.spines["right"].set_visible(False)    

# Ensure that the axis ticks only show up on the bottom and left of the plot.    
ax.get_xaxis().tick_bottom()    
ax.get_yaxis().tick_left()

plt.xlabel("Proportion of the population that is sick.")
plt.ylabel("Probability that a person is sick given that they test positive.")


for p_pws in np.arange(0.9, 1, 0.01):
    x = np.arange(0.00001, 0.10, 0.00001)
    y = p_sick_when_pos(x, p_pws, p_pwh) 
    plt.legend(str(round(p_pwh,3)))
    plt.plot(x,y, label = str(round(p_pws,3)));
ax.legend(title = "Probability that a sick persons tests positive")
plt.title("Effectiveness of a test when the probability" + 
          " that a healthy person tests positive is " + str(p_pwh) + 
         "for different values of the probability that a sick person tests positive.");
```
